# Handling Missing Value and Duplicated Value

# Data Type Conversion

```python
# convert date column to datetime
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df['Product Name']=df['Product Name'].astype(str).str[:20]+"..."
```

# Remove Useless columns

```python
# remove useless columns
columns=["Row ID","Customer ID","Customer Name","Postal Code"]

df.drop(columns=columns,axis=1,inplace=True)
```

# UNIVARIATE ANALYSIS

```python
# Univariate Analysis divide into two parts uni(one) and variate(variable) means we can say one variable column analysis
print("\n === CATEGORICAL COLUMN ANALYSIS ===\n")

cat_cols=["Segment","Region","Category"]
color=["red","green","skyblue","purple"]

for col in cat_cols:
    print(f"\n --- {col} ---\n")
    print(df[col].value_counts())

    # visualization
    plt.figure(figsize=(5,4))
    df[col].value_counts().plot(kind="bar",color=np.random.choice(color))
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

##### I performed univariate analysis to understand each feature individually by checking its distribution using value_counts() and visualizing it with bar charts
##### • Segment:
##### “Consumer segment has the highest frequency, indicating it is the main customer base of the business.”
##### • Category:
##### “Office Supplies category appears most frequently, suggesting it has the highest demand or sales volume.”
##### • Region:
##### “West region has the highest count, showing that most business activity or transactions are concentrated in this region.”

# Numerical Column Analysis

```python
# numerical column analysis
print("\n === NUMERICAL COLUMN ANALYSIS === \n")

columns=["Sales","Quantity","Discount","Profit"]

for col in columns:
    print(f"\n --- {col} --- \n")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Std: {df[col].std()}")
    print(f"Min: {df[col].min()}")
    print(f"Max: {df[col].max()}\n")

    # visualization
    plt.figure(figsize=(7,3))

    plt.subplot(121)
    df[col].hist(bins=30,color="lightgreen")
    plt.title(f"Distribution of {col}")

    plt.subplot(122)
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
    
```

##### For numerical features, I calculated summary statistics like mean, median, standard deviation, min and max. I also used histograms and boxplots to understand distribution and detect outliers.
##### • Sales data is highly right-skewed with most transactions having low values, while a few high-value transactions act as outliers.
##### • Quantity is concentrated around lower values, indicating most orders are small, with few bulk orders as outliers.
##### • Discount is mostly low, with majority of transactions having minimal discounts, while high discount cases are less frequent.
##### • Profit shows both positive and negative values, indicating presence of loss-making transactions, with significant outliers on both ends

# BIVARIATE ANALYSIS

```python
# Bivariate Analysis is sa statistical method used to analyze the relationship between two variables and to determine how one variable affects or is associated with another

print("\n---------------------SALES & PROFIT BY CATEGORY------------------------\n")

category_performance=df.groupby('Category').agg({
    "Sales": 'sum',
    "Profit": 'sum',
    "Quantity": 'sum',
    "Discount":'mean',
    "Order ID": 'count'
}).rename(columns={"Order ID": "Order Count"})

# calculate profit margin
category_performance["Profit Margin"]=(category_performance["Profit"]/category_performance["Sales"])*100

print(category_performance,"\n\n")

# visualization
fig,axes=plt.subplots(3,2,figsize=(18,12))

# Sales by Category
category_performance['Sales'].plot(kind="bar",ax=axes[0,0],color='blue')
axes[0,0].set_title("Sales by Category")
axes[0,0].tick_params(axis='x',rotation=30)

# Profit by category
category_performance['Profit'].plot(kind="bar",ax=axes[0,1],color='green')
axes[0,1].set_title("Profit by Category")
axes[0,1].tick_params(axis='x',rotation=30)

# Order count by Category
category_performance["Order Count"].plot(kind="bar",ax=axes[1,0],color='red')
axes[1,0].set_title("Order Count by Category")
axes[1,0].tick_params(axis='x',rotation=30)

# Profit margin by Category
category_performance["Profit Margin"].plot(kind='bar',ax=axes[1,1],color="skyblue")
axes[1,1].set_title("Profit Margin by Category")
axes[1,1].tick_params(axis='x',rotation=30)

# quantity sold by category
category_performance["Quantity"].plot(kind='bar',ax=axes[2,0],color='pink')
axes[2,0].set_title("Total Quantity sold by Category")
axes[2,0].tick_params(axis='x',rotation=30)

# avg discount by category
category_performance["Discount"].plot(kind='bar',ax=axes[2,1],color='orange')
axes[2,1].set_title("Average Discount by Category")
axes[2,1].tick_params(axis='x',rotation=30)
plt.tight_layout()
plt.show()
```

##### I performed bivariate analysis to compare categories like metrics sales,profit,quantity, and discount to understand their impact on business performance
##### Technology drives revenue and profit through higher-value products, Office Supplies contributes via high-volume sales, while higher discounting in Furniture suggests an effort to boost demand but may be impacting profitability

# Regional Performance

```python
# regional performance

print(f"----------Sales & Profit by Region-------------")
region_performance=df.groupby('Region').agg({
    "Sales":'sum',
    "Profit": 'sum',
    "Quantity":'sum',
    "Discount":'mean',
    "Order ID": 'count'
}).rename(columns={'Order ID': "Order Count"})

region_performance['Profit Margin']=(region_performance["Profit"]/region_performance["Sales"])*100
print(region_performance,"\n\n")

# visualization
plt.figure(figsize=(18,9))

# sales by region
plt.subplot(321)
region_performance['Sales'].plot(kind='pie',autopct="%.2f%%")

# profit by region
plt.subplot(322)
region_performance['Profit'].plot(kind="pie",autopct="%.2f%%")

# Order count by region
plt.subplot(323)
region_performance["Order Count"].plot(kind='bar',color='red')
plt.title("Region wise Order count")
plt.xticks(rotation=30)
plt.xlabel("Region")
plt.ylabel("Order Count")

# Profit Margin by region
plt.subplot(324)
region_performance["Profit Margin"].plot(kind='bar',color='green')
plt.title("Region wise Profit Margin")
plt.xticks(rotation=30)
plt.xlabel("Region")
plt.ylabel("Profit Margin")

# Total sold quantity by region
plt.subplot(325)
region_performance["Quantity"].plot(kind='bar',color='pink')
plt.title("Region wise Total Quantity Sold")
plt.xticks(rotation=30)
plt.xlabel("Region")
plt.ylabel("Quantity")

# Average Discount by region
plt.subplot(326)
region_performance["Discount"].plot(kind='bar',color='orange')
plt.title("Average Discount vs Profit ,Region wise")
plt.xticks(rotation=30)
plt.xlabel("Region")
plt.ylabel("Discount")

plt.tight_layout()
plt.show()
```

#### West region is the strongest performer, with high sales and profit due to strong demand and lower discounts, while higher discounts in the Central region may be reducing profitability.The business should continue to strengthen its presence in the West, and in the Central region, focus on reducing excessive discounts and improving sales strategies to increase profitability.

# Time Based  Analysis 

```python
# TIme series analysis
import random
print("\n === TIME SERIES ANALYSIS === \n")

df['Order month']=df['Order Date'].dt.month_name()
df['Order num']=df['Order Date'].dt.month
df['Order year']=df['Order Date'].dt.year

# Order monthe wise sales & profit
order_month=df.groupby(['Order month','Order num']).agg({
    "Sales": 'sum',
    "Profit":'sum',
    "Quantity": 'sum'
}).reset_index().sort_values(by="Order num",ascending=True)

# find profit margin from data
order_month["pct_growth"]=order_month["Sales"].pct_change()
order_month["Profit Margin"] =(order_month["Profit"]/order_month["Sales"])*100
print(order_month,"\n\n")

#Visualization
colors=["red","green","grey","brown","pink","skyblue","blue"]
#function for finding month wise sales,profit ......
def monthly_sales(col1,col2):
    sns.lineplot(x=col1,y=col2,data=order_month,marker='o',color=random.choice(colors))
    plt.title(f"Order month wise {col2}")
    plt.grid(True)
    plt.tight_layout()

plt.figure(figsize=(11,10))

    
# Monthly Sales
plt.subplot(511)
monthly_sales("Order month","Sales")

#Monthly Profit
plt.subplot(512)
monthly_sales("Order month","Profit")

# Monthly sold Quantity
plt.subplot(513)
monthly_sales("Order month","Quantity")


plt.tight_layout()
plt.show()
```

#### Sales and quantity peak in November, indicating a surge in demand, likely due to seasonal or festive factors, while profit is highest in December, suggesting better margins or reduced discounting during that period.
#### The business should prepare for high demand in November and focus on maximizing profitability in December through optimized pricing strategies

# Trend Anaylsis month-on-month

```python
# Trend Analysis month on month

print("-------------Trend Analysis----------------")
pct_growth=order_month["Sales"].pct_change()*100
plt.figure(figsize=(12,5))
plt.plot(order_month["Order month"],pct_growth,marker='o',color='blue')
plt.grid(True)
plt.xlabel("Month")
plt.ylabel("Growth Percentage")
plt.title("Tree Analysis month-on-month")
```

#### Month-to-month performance shows ups and downs, with major growth in March, September, and November, while some months show decline, indicating fluctuating demand patterns.

# Key Insights & Summary

```python
print("\n === Key Insights & Summary === \n")

print("--------Overall Performance----------\n")

total_sales=df["Sales"].sum()
total_profit=df['Profit'].sum()
total_sold_quantity=df['Quantity'].sum()
order_count=df["Order ID"].nunique()
avg_profit_margin=(total_profit/total_sales)*100

print("=== Summary Report & KPI === \n")

print(f"Total Sales: {total_sales}")
print(f"Total Profit: {total_profit}")
print(f"Total sold_quantity : {total_sold_quantity}")
print(f"Total Order count : {order_count}")
print(f"Total Average Profit Margin: {avg_profit_margin}")

# Top performing categories
top_categories=df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f"\nTOP PERFORMING CATEGORIES\n")
for i,(category,profit) in enumerate(top_categories.items()):
    print(f"{i+1}. {category}: ${profit: .2f}")

# top perforing region
top_region=df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(f"\nTOP PERFORMING REGION\n")
for i,(region,profit) in enumerate(top_region.items()):
    print(f"{i+1}.{region}: {profit:.2f}")

# Top segment performance
top_segment=df.groupby('Segment')["Profit"].sum().sort_values(ascending=False)
print(f"\nTOP PERFORMING SEGMENT\n")
for i,(segment,profit) in enumerate(top_segment.items()):
    print(f"{i+1}.{segment}: {profit: .2f}")

# Best region,category and segment
print("\nINSIGHT CATEGORY\n")
best_category=top_categories.idxmax()
worst_category=top_categories.idxmin()
print(f"{best_category} is the most profitable category, indicating strong customer demand and higher margins. while {worst_category} category underperforms, which may be due to high discounts or lower profit margins")

print("\nINSIGHT REGION\n")
best_region=top_region.idxmax()
worst_region=top_region.idxmin()
print(f"{best_region} generates the highest profit, suggesting a strong customer base and better sales performance.{worst_region} shows lower profit, which may require improved marketing or pricing strategies.")

print("\nINSIGHT SEGMENT\n")
best_segment=top_segment.idxmax()
worst_segment=top_segment.idxmin()
print(f"{best_segment} segment contributes the highest profit, indicating high demand from individual customers.{worst_segment} segment underperforms and may need better targeting or promotional strategies")
```
